```python
fund_master = pd.read_csv(r"C:\Users\swapnil\Mutual Fund Analytics\data\raw\01_fund_master.csv")
nav_history = pd.read_csv(r"C:\Users\swapnil\Mutual Fund Analytics\data\raw\02_nav_history.csv")
aum_fund_house = pd.read_csv(r"C:\Users\swapnil\Mutual Fund Analytics\data\raw\03_aum_by_fund_house.csv")
monthly_sip = pd.read_csv(r"C:\Users\swapnil\Mutual Fund Analytics\data\raw\04_monthly_sip_inflows.csv")
category_inflows = pd.read_csv(r"C:\Users\swapnil\Mutual Fund Analytics\data\raw\05_category_inflows.csv")
industry_folio = pd.read_csv(r"C:\Users\swapnil\Mutual Fund Analytics\data\raw\06_industry_folio_count.csv")
scheme_performance = pd.read_csv(r"C:\Users\swapnil\Mutual Fund Analytics\data\raw\07_scheme_performance.csv")
investor_transactions = pd.read_csv(r"C:\Users\swapnil\Mutual Fund Analytics\data\raw\08_investor_transactions.csv")
portfolio_holdings = pd.read_csv(r"C:\Users\swapnil\Mutual Fund Analytics\data\raw\09_portfolio_holdings.csv")
benchmark_indices = pd.read_csv(r"C:\Users\swapnil\Mutual Fund Analytics\data\raw\10_benchmark_indices.csv")
```
