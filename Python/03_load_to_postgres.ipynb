```python
# Import necessary libraries
from sqlalchemy import create_engine
from sqlalchemy.exc import OperationalError
import sys

# Method 1: Basic connection with error handling
try:
    # Replace these with your actual database credentials
    username = "postgres"
    password = "swapnil@123" 
    host = "localhost"  # Changed from "123@localhost" to just "localhost"
    port = "5432"
    database = "your_database_name"
    
    # Construct proper connection string
    DATABASE_URL = f"postgresql://{username}:{password}@{host}:{port}/{database}"
    
    # Create engine
    engine = create_engine(DATABASE_URL)
    
    # Test connection
    connection = engine.connect()
    print("Connected Successfully!")
    connection.close()
    
except OperationalError as e:
    print(f"Connection failed: {e}")
    print("\nTroubleshooting steps:")
    print("1. Check if PostgreSQL server is running")
    print("2. Verify username, password, and database name")
    print("3. Ensure the database exists")
    print("4. Check if host and port are correct")

# Method 2: Alternative connection without credentials (for local SQLite)
try:
    # For testing purposes, use SQLite instead
    engine_sqlite = create_engine("sqlite:///test.db")
    connection_sqlite = engine_sqlite.connect()
    print("SQLite connection successful!")
    connection_sqlite.close()
except Exception as e:
    print(f"SQLite connection failed: {e}")
```
